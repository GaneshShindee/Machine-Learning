#1. Introduction to Pandas
###DataFrames and Series
Pandas provides two main data structures: Series and DataFrame. A Series is a one-dimensional array-like object, and a DataFrame is a two-dimensional, size-mutable, and potentially heterogeneous tabular data structure.

Example: Creating a Series and DataFrame

```python
import pandas as pd

# Create a Series
series = pd.Series([1, 2, 3, 4, 5])
print("Series:\n", series)

# Create a DataFrame
data = {'A': [1, 2, 3], 'B': [4, 5, 6], 'C': [7, 8, 9]}
df = pd.DataFrame(data)
print("DataFrame:\n", df)
```

###Basic DataFrame Operations
DataFrames allow you to perform various operations such as selection, filtering, and modification.

Example: Selecting and Filtering Data

```python
# Select a column
print("Column A:\n", df['A'])

# Select multiple columns
print("Columns A and B:\n", df[['A', 'B']])

# Filter rows
filtered_df = df[df['A'] > 1]
print("Filtered DataFrame:\n", filtered_df)
```

###Data Cleaning
Data cleaning involves handling missing values and correcting or removing inaccurate records.

Example: Handling Missing Values

```python
# Create a DataFrame with missing values
data_with_nan = {'A': [1, 2, None], 'B': [4, None, 6], 'C': [None, 8, 9]}
df_nan = pd.DataFrame(data_with_nan)
print("DataFrame with NaNs:\n", df_nan)

# Fill missing values with the mean
df_filled = df_nan.fillna(df_nan.mean())
print("DataFrame with NaNs filled:\n", df_filled)

# Drop rows with missing values
df_dropped = df_nan.dropna()
print("DataFrame with NaNs dropped:\n", df_dropped)
```

###Data Aggregation and Grouping
Pandas allows you to aggregate and group data to perform summary statistics.

Example: Grouping and Aggregating Data

```python
# Create a DataFrame
data = {'Category': ['A', 'A', 'B', 'B'], 'Value': [10, 15, 10, 20]}
df = pd.DataFrame(data)

# Group by 'Category' and calculate the mean
grouped_df = df.groupby('Category').mean()
print("Grouped DataFrame:\n", grouped_df)
```

#2. Introduction to Matplotlib
###Basic Plotting
Matplotlib is a powerful library for creating static, animated, and interactive visualizations in Python.

Example: Line Plot

```python
import matplotlib.pyplot as plt

# Data
x = [1, 2, 3, 4, 5]
y = [2, 3, 5, 7, 11]

# Create a line plot
plt.plot(x, y)
plt.title('Line Plot')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.show()
```

###Customizing Plots
You can customize plots with titles, labels, legends, and more.

Example: Customized Plot

```python
# Create a line plot with customizations
plt.plot(x, y, marker='o', linestyle='--', color='r', label='Prime Numbers')
plt.title('Customized Line Plot')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.legend()
plt.grid(True)
plt.show()
```

###Subplots
Matplotlib allows you to create multiple plots in a single figure using subplots.

Example: Subplots

```python
# Create subplots
fig, axs = plt.subplots(2, 2)

# First subplot
axs[0, 0].plot(x, y)
axs[0, 0].set_title('Plot 1')

# Second subplot
axs[0, 1].scatter(x, y)
axs[0, 1].set_title('Plot 2')

# Third subplot
axs[1, 0].bar(x, y)
axs[1, 0].set_title('Plot 3')

# Fourth subplot
axs[1, 1].hist(y)
axs[1, 1].set_title('Plot 4')

plt.tight_layout()
plt.show()
```

#3. Introduction to Seaborn
##Basic Plotting with Seaborn
Seaborn is a Python visualization library based on Matplotlib that provides a high-level interface for drawing attractive statistical graphics.

Example: Scatter Plot

```python
import seaborn as sns

# Load dataset
tips = sns.load_dataset('tips')

# Create a scatter plot
sns.scatterplot(x='total_bill', y='tip', data=tips)
plt.title('Scatter Plot')
plt.show()
```

###Statistical Plots
Seaborn provides functions for visualizing statistical relationships.

Example: Pair Plot

```python
# Create a pair plot
sns.pairplot(tips)
plt.title('Pair Plot')
plt.show()
```

###Example: Box Plot


```python
# Create a box plot
sns.boxplot(x='day', y='total_bill', data=tips)
plt.title('Box Plot')
plt.show()
```

###Customizing Seaborn Plots
Seaborn plots can be customized with themes, color palettes, and more.

Example: Customized Seaborn Plot

```python
# Set the theme
sns.set_theme(style="darkgrid")

# Create a bar plot with a custom color palette
sns.barplot(x='day', y='total_bill', data=tips, palette='viridis')
plt.title('Customized Bar Plot')
plt.show()
```
